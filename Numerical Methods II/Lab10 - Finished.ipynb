```python
A = array([[-33, 36, -27, 18, -9], [-33, 51,-45, 30, -15], [ -19, 38, -45, 34, -17], [ -9, 18, -27, 27, -15],
           [ -3, 6, -9, 12, -9]])
print("A = ", A)
B = array([[120], [186], [162], [99], [36]])
print("Segundo miembro: B = ", B)
exito, X = gauss_pp_verbose(A, B)
if exito:
    print("Solución: X = ", X)
    print("Comprobación: ||B-A@X||_2 = ", norma_vec(B-A@X, 2))
else:
    print("Error ", X)
```

Tomando como vector inicial aquel que tiene todas sus componentes igual a 1, y aplicando el método de la potencia, calcular su autovalor de módulo mayor y menor; calcular también cond2(𝐴). Ídem para $𝑛=20$.
    </span></div>

```python
A = 4*eye(5) - eye(5, k=-1) - eye(5, k=-2) - eye(5, k=2) - eye(5, k=1)
print("A = ", A)
X_0 = array([[1], [1], [1], [1], [1]])
print("X_0 = ", X_0)

exito, normas1, lambdas, X = potencia(A, X_0, inf, 200, 1e-12)
if exito:
    print('Convergencia de las normas: ', normas1)
    print('Convergencia de los cocientes:', lambdas)   
    print('Convergencia de los vectores: ', X)
else:
    print(normas)
    
exito, normas2, lambdas, X = potenciainv(A, X_0, inf, 200, 1e-12)
if exito:
    print('Convergencia de las normas: ', normas2)
    print('Convergencia de los cocientes:', lambdas)    
    print('Convergencia de los vectores: ', X)
else:
    print(normas)


#A es normal por ser simétrica
cond_2_A = normas1*normas2
print("Cond_2(A) = ", cond_2_A)
```
