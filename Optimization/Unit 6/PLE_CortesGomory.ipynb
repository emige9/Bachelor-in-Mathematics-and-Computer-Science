$$
\begin{array}{rrcl}
    \min & 3x_{1} +4x_{2}&\\
    {\rm sujeto\  a}& x_1 + 2x_2 & \geq& 4\\
    & 3x_1 + x_2 &\geq& 4\\
    &x_1, x_2 & \geq & 0\\
    &x_1, x_2 & {\rm enteras}
\end{array}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, IntSlider

def plot_line(k):
    plt.figure(figsize=(10, 6))
    plt.xlim(-2, 8)  # Ajusta los lmites para que coincidan con la imagen
    plt.ylim(-5, 10)

    x = np.linspace(-2, 8, 100)

    # Ecuaciones de las rectas
    y_rojo = (4 - x)/2
    y_azul = 4 - 3*x
    y_verde = (k - 3*x) / 4

    # Graficar las rectas
    plt.plot(x, y_rojo, 'r-', label='$x_1 + 2x_2 = 4$')
    plt.plot(x, y_azul, 'b-', label='$3x_1 + x_2 = 4$')
    plt.plot(x, y_verde, 'g--', label=f'$3x_1 + 4x_2 =$ {k}')

    # Encontrar puntos enteros dentro del polgono
    def is_inside(x, y):
        return x >= 0 and y >= 0 and y >= (4-x)/2 and y >= 4 - 3*x

    points_inside = []
    for x in range(0, 8):
        for y in range(0, 10):
            if is_inside(x, y):
                points_inside.append((x, y))

    # Graficar puntos
    x_inside, y_inside = zip(*points_inside)
    plt.scatter(x_inside, y_inside, color='red')

    # Marcar puntos especficos y agregar texto
    plt.scatter(2, 1, color='black', marker='o')
    plt.text(2, 1.3, "Solucin ptima de PE: (2,1)")
    plt.scatter(0.8, 1.60, color='black', marker='o')
    plt.text(-0.5, 0.8, "Solucin ptima de PR ")
    plt.text(0, 0.3, "(4/5, 8/5)")

    # Personalizar grfico
    plt.title('Redondeo en Programacin Entera')
    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    plt.legend()
    plt.grid(True)
    plt.show()

# Crear un slider interactivo para k
interact(plot_line, k=IntSlider(min=-5, max=15, step=1, value=4))
```
